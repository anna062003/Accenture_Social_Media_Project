```python
import pandas as pd
import sqlite3

#read the CSV
df_content = pd.read_csv('Content.csv')
df_reaction = pd.read_csv('Reactions.csv')
df_re_type = pd.read_csv('ReactionTypes.csv')

#connect to a database
conn = sqlite3.connect("content.db")
conn1 = sqlite3.connect("Reactions.db")
conn2 = sqlite3.connect("ReactionTypes.db")

#if the db does not exist, this creates a Any_Database_Name.db file in the current directory
#store your table in the database:
df_content.to_sql('content', conn)
df_reaction.to_sql('reactions', conn1)
df_re_type.to_sql('reactionsType',conn2)

#read a SQL Query out of your database and into a pandas dataframe
sql_string = 'SELECT * FROM Content'
df_content = pd.read_sql(sql_string, conn)
df_reaction = pd.read_sql(sql_string, conn1)
df_re_type = pd.read_sql(sql_string, conn2)
```

```python
df_mer = df_reaction.merge(df_re_type, how='inner', on = 'Type')
df_mer = df_mer[["Content ID", "Type", "Sentiment", "Score", "Datetime"]]
df_mer
```

```python
df_merge = df_mer.merge(df_content, how = 'inner', on = ["Content ID"])
df_merge = df_merge[["Content ID", "Category", "Type_y", "Type_x", "Score", "Datetime"]]
df_merge = df_merge.rename(columns = {"Type_y": "Content Type", "Type_x": "Reaction Type"})
df_merge["Category"] = df_merge["Category"].str.strip('"')
df_merge["Category"] = df_merge["Category"].str.lower()
df_merge
```

```python
df_merge.to_csv("Final_content_result.csv")
```

```python
df_merge.count()
```

```python
df_merge_cal = df_merge.groupby("Category").agg({"Score": "sum"})
df_merge_cal = df_merge_cal.sort_values(by="Score", ascending = False)
df_merge_cal.head()
```
